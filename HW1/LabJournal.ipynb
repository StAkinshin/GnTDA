```python
# команды выполнялись в терминале Ubuntu (WSL2)
#все пакеты устанавливались через conda
```

```python
# Шаг №1: скачивание данных
# данные в файле spades_scaffolds.fasta, файл был скопирован в рабочую директорию
```

```python
# Шаг №2: оценка качества сборки - QUAST
# инструмент QUAST (версия 5.3.0)
# результаты были перенесены в отчёт
quast spades_scaffolds.fasta -o quast_out
```

```python
# Шаг №3: отбор контигов по длине более 3000 пар оснований
more spades_scaffolds.fasta.fai | awk '$2 >= 3000 {print $1}' | xargs samtools faidx spades_scaffolds.fasta > 3000.fa
grep '^>' 3000.fa # получили 23 контига, которые удовлетроряли по длине
# Контиг NODE_1 имеет длину 29 907 bp и покрытие 150х - искомый, контиг NODE_3 с чересчур высоким покрытием (>13 000) и длиной 5584 bp - скорее всего, это технические остатки от секверирования (бактериофаг)
```

```python
# Шаг №4: последовательность NODE_1 отправили на сайт BLAST для выравнивания с вирусами
# использовали blastn, дату ограничили 1 января 2020 года (1900/01/01:2020/01/01[PDAT])
# найденный контиг имел идентичность порядка 89% с изолятом коронавируса летучих мышей SL-CoVZC45
# результаты после анализа BLAST были записаны в отчет
```
